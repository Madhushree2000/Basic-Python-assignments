```python
import pandas as pd
from numpy import *
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.data')
X = array(df)[:,0:4]
Y = array(df)[:,4:]
df.head()
```

```python
def target(Lable):

    A = []
    output = []
    x = 0

    for i in Lable:
        if (i not in A):
            A.append(i)
        
    for i in Lable:
        x = A.index(i)
        output.append(x)
        
    return(array(output))
```

```python
Y = target(Y)
Y
```

```python
def logsig(net):
    l = 1 + exp(-net)
    return(1/l)
```

```python

global Weight;
global Bias;
global MSE;

Weight = []
Bias = []
MSE = []
       
```

```python
def neuron(W,B,X,Y,n):

        net = 0
        Error = []

        #1. Calculating the net value
        for i in range(0,len(X)):
                net = net + X[i] * W[i] + B
        
        #2. Transfer function
        ao = logsig(net)

        #3. Error calculation
        er = Y - ao
        Error.append(er)

        #Weight and bias modification
        if(er > 0):
                W = W + er * X
                Weight[n] = W
                B = B + er
                Bias[n] = B


        if(er < 0):

                W = W - er * X
                Weight[n] = W
                B = B - er
                Bias[n] = B
                
        if(er == 0):
            pass

        me = (matrix(Error) * (matrix(Error).T)).tolist()[0][0]

        return me
```

```python

for j in range(0,len(X)):
    W = random.rand(4)
    B = random.rand()
    Weight.append(W)
    Bias.append(B)

plt.ion()
plt.figure(1)

for i in range(0,50):

    for j in range(0,len(X)):
        W = Weight[j]
        B = Bias[j]
        X1 = X[j]
        Y1 = Y[j]
        m = neuron(W,B,X1,Y1,j)

    MSE.append(m)
    plt.plot(MSE, 'r.-')
    plt.pause(0.001)
    plt.draw()

plt.show()
        
        
```
