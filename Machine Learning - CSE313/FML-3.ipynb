```python
from numpy import *
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
```

```python
def minfunc(line):
    m = 100
    for x in line:
        if (x < m):
            m = x
            
    return m
```

```python
def maxfunc(line):
    m = 0
    for x in line:
        if (x > m):
            m = x
            
    return m
```

```python
def meanfunc(line):
    sum = 0
    
    for x in line:
        sum = sum + x
        
    m = sum / len(line)
    
    return m
```

```python
def stdfunc(line):
    sum = 0
    
    for x in line:
        sum = sum + x
    
    mean = sum/len(line)

    for x in line:
        sum = sum + square(x - mean)

    std = sqrt(sum/len(line))
    return std                              
```

```python
iris = pd.read_csv('iris.data')
iris = iris.to_numpy()
iris = delete(iris,-1,1)
minlist = []
maxlist = []
stdlist = []
meanlist = []
        
for x in iris:
            
            min = minfunc(x)
            minlist.append(min)
            
            max = maxfunc(x)
            maxlist.append(max)
            
            mean = meanfunc(x)
            meanlist.append(mean)
            
            std = stdfunc(x)
            stdlist.append(std)
```

```python
plt.subplot()
plt.plot(minlist, c='b')
plt.title ("minimum")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.subplot()
plt.plot(maxlist, c='g', linestyle='--')
plt.title ("maximum")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.subplot()
plt.plot(stdlist, c='#00FFFF', linestyle='-.')
plt.title ("Standard Deviation")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.subplot()
plt.plot(stdlist, c='#FF0040', linestyle=':')
plt.title ("Standard Deviation")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.subplot()
plt.scatter(meanlist,stdlist, c = "#FE642E")
plt.xlabel('Mean')
plt.ylabel('Standard Deviation')
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(meanlist, minlist, maxlist, c = '#FA5882')
ax.set_xlabel('mean')
ax.set_ylabel('min')
ax.set_zlabel('max')
plt.show()
```
