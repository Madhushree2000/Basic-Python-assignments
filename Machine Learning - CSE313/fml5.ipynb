```python
import os, pdb
import pandas as pd
from numpy import *
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv('iris.csv')
```

```python
p2 = wine.iloc[:,1:5].to_numpy()
t = wine.iloc[:,0].to_numpy()
```

```python
def hardlim(n):
    
    if n < 0:
        return -1
        
    else:
        return 1
```

```python
def singleNeuron(xx, t):
    
    #1. net = wp
    
    ww = [0,0,0,0]
    b = matrix([0])
    weight = matrix(ww)
    temp_err = []
    for pat, tar in zip(xx, t):
        
        pat = matrix(pat)
        tar = matrix(tar)
        net = weight*pat.T +b 
        
        #2. a = f(net) 
        ao = hardlim(net)
        
        #3. Error = t-yin
        error = (tar - net).tolist()[0][0]
        temp_err.append(error)
        
        #modify the weight
        alpha = 1
        
        if error > 0:
            weight = weight + alpha*error*pat
            b = b + alpha*error
        elif error < 0:
            weight = weight + alpha*error*pat
            b = b + alpha*error
        else:
            weight = weight
            b = b

        print("\nWeight=", weight)
        
    return temp_err
```

```python
sse_list = []
plt.figure('1')
plt.ion

for i in range(0,100):
    error_list = singleNeuron(p2, t)
    e1 = (matrix(error_list) * matrix(error_list).T).tolist()[0][0]
    sse_list.append(e1)
    print(e1)
    plt.subplot(111)
    
    plt.plot(sse_list)
    plt.grid(True)
    plt.pause(0.01)
    plt.draw()

```
