```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 
```

```python
ecg = pd.read_csv('ECG Data.csv')
```

```python
min = []
max = []
mean = []
std = []

def minimum_function(row):
    min = row[0]
    for i in row:
        if i<min:
            min = i
    return min

def maximum_function(row):
    max = row[0]
    for i in row:
        if i>max:
            max = i
    return max

def mean_function(row):
    sum = 0
    for i in row:
        sum = sum + i
    mean = sum/len(row)
    return mean

def std_function(row):
    sum = 0
    for i in row:
        sum = sum + i
    xbar = sum/len(row)

    for x in row:
        sum = sum + np.square(x - xbar)

    std = np.sqrt(sum/len(row))
    return std
```

```python
window_dict = {}
min_dict = {}
max_dict = {}
mean_dict = {}
std_dict = {}
for i in range (0,len(ecg.columns),10):
    window = ecg.iloc[:,i:i+10]
    window = window.to_numpy()
    window_dict['window_'+ str(i)] = window
    for j in window:
        min.append(minimum_function(j))
        max.append(maximum_function(j))
        mean.append(mean_function(j))
        std.append(std_function(j))
    
    min_dict['min' + str(i)] = min
    max_dict['max' + str(i)] = max
    mean_dict['mean' + str(i)] = mean
    std_dict['std' + str(i)] = std


```

```python
plt.figsize = (8,6)
plt.subplot(1,2,1)
plt.hist(min_dict['min0'], alpha = 0.5, label= 'Window 1',color= 'y')
plt.legend(loc='upper right')
plt.xlabel("Data", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Histogram of Minimum Values")
plt.subplot(1,2,2)
plt.hist(min_dict['min180'], alpha = 0.5, label= 'Window 18')
plt.legend(loc='upper right')
plt.xlabel("Data", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Histogram of Minimum Values")
plt.subplots_adjust(wspace=0.8, 
                    hspace=0.8)
plt.show()
```

```python
plt.figsize = (8,6)
plt.hist(std_dict['std20'], alpha = 0.5, label= 'Window 3')
plt.hist(std_dict['std30'], alpha = 0.5, label= 'Window 4')
plt.legend(loc='upper right')
plt.xlabel("Data", size=14)
plt.ylabel("Frequency", size=14)
plt.title("Histogram of Standard Deviation")
plt.show()
```

```python
L = ['mean0', 'mean10', 'mean20','mean30',]
n = [mean_dict[i]for i in L]
n_bin = 10
plt.hist(n, n_bin, histtype = 'bar', stacked = True, label= [i for i in L])
plt.legend(loc='upper right')
plt.show()
```
